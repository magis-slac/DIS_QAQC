```python
import imageio
import math
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
f_numbers = np.array([
    1.4, 1.6, 2.0, 2.8, 4.0, 5.6, 8.0
])

high_threshold_dict = {
    60: np.array([1.491, 1.413, 1.824, 2.638, 3.653, 4.977, 7.122]),
    62: np.array([1.330, 1.408, 1.774, 2.838, 3.636, 5.393, 7.250]),
    65: np.array([1.219, 1.419, 1.873, 2.767, 3.786, 5.532, 7.521])
}

low_threshold_dict = {
    60: np.array([2.101, 1.957, 2.261, 3.198, 4.268, 6.069, 8.309]),
    62: np.array([1.746, 1.807, 2.178, 3.281, 4.379, 6.380, 8.630]),
    65: np.array([1.752, 1.879, 2.278, 3.165, 4.340, 6.435, 9.062])
}
```

```python
for WD, DoF in high_threshold_dict.items():
    plt.plot(f_numbers, DoF,
             marker='s', ls='--', label=f"WD={WD}mm, High-th.")
    
plt.gca().set_prop_cycle(None)
    
for WD, DoF in low_threshold_dict.items():
    plt.plot(f_numbers, DoF,
             marker='^', ls='-.', label=f"WD={WD}mm, Low-th.")

f_num_min, f_num_max = 1.2, 8.2

target_dof_1 = 4
plt.axhline(target_dof_1,
            ls='-', color='red', alpha=0.3)
plt.arrow(7.5, target_dof_1 + 0.1, 0, 0.5,
          width=0.08, color='red', alpha=0.5)

target_dof_2 = 2.5
plt.axhline(target_dof_2,
            ls='-', color='blue', alpha=0.3)
plt.arrow(7.5, target_dof_2 + 0.1, 0, 0.5,
          width=0.08, color='blue', alpha=0.5)

plt.xlabel("f/#")
plt.xlim([f_num_min, f_num_max])
plt.ylabel("DoF [mm]")
plt.ylim([0, 10])
plt.grid(True)
plt.legend()
plt.show()
```
